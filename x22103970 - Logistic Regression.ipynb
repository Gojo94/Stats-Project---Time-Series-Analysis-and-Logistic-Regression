#### Visualization of the Data

```r
library(ggplot2)

histogram <- function(x)
{
title <- paste(deparse(substitute(x), 500), collapse="\n")
sdx <- sd(x)
mx <- mean(x)
hist(x, prob=TRUE,
main=paste("Histogram of ",title),
xlim=c(mx-3*sdx, mx+3*sdx), ylim=c(0, 0.5/sdx))
curve(dnorm(x, mean=mx, sd=sdx), col='red', lwd=3, add=TRUE)
}



par(mfrow=c(2,2))
histogram(data1$AGE)
plot.new() # placeholder
histogram(data1$Urea)
histogram(log(data1$Urea))
histogram(data1$Cr)
histogram(log(data1$Cr))
histogram(data1$HbA1c)
histogram(log(data1$HbA1c))
histogram(data1$Chol)
#histogram(log(data1$Chol))
plot.new() # placeholder
histogram(data1$TG)
histogram(log(data1$TG))
histogram(data1$HDL)
histogram(log(data1$HDL))
histogram(data1$LDL)
histogram(log(data1$LDL))
histogram(data1$VLDL)
histogram(log(data1$VLDL))
histogram(data1$BMI)
histogram(log(data1$BMI))
#histogram(data$CLASS)

ggplot(data = data1, aes(x = Gender)) +
  geom_bar(fill = "steelblue") +
  labs(title = "Distribution of Gender")
```

#### Transform the variable

```r
# Transform the variables
data1$Urea <- log(data1$Urea)
data1$Cr <- log(data1$Cr)
data1$TG <- log(data1$TG)
data1$HDL <- log(data1$HDL)
data1$LDL <- log(data1$LDL)
data1$VLDL <- log(data1$VLDL)
data1$BMI <- log(data1$BMI)
```

#### Split the dataframe

```r
#install.packages("caTools")

# Set the seed for reproducibility
set.seed(25)

# Split the data into training and test sets
library(caTools)

split <- sample.split(data1$CLASS, SplitRatio = 0.7)
train <- subset(data1, split == TRUE)
test <- subset(data1, split == FALSE)
```

### Building the model

#### First model

```r
model <- glm(CLASS ~ ., data = train, family = "binomial")
summary(model)
par(mfrow=c(2,2))
plot(model)
```

```r
#Evaluate the Model
test_prediction <- predict(model, newdata=test, type = 'response')

cm <- table(test$CLASS, test_prediction > 0.5)

accuracy <- sum(diag(cm) / sum(cm))

accuracy
round(accuracy, 4) * 100
```
